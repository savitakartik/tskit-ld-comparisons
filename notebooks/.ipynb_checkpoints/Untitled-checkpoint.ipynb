```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df_w10 = pd.read_csv("calc_times_w10.csv")
df_w10 = df_w10.sort_values(['sample_size'], ascending=[True])
df_w10
```

```python
df_w100 = pd.read_csv("calc_times_w100.csv")
df_w100 = df_w100.sort_values(['sample_size'], ascending=[True])
df_w100
```

```python
plt.plot(df_w10['sample_size'], df_w10['plink_time_taken'], label="plink: window size 10", marker='o', color="red")
plt.plot(df_w10['sample_size'], df_w10['tskit_time_taken'], label="tskit: window size 10", marker='o', color="blue")
plt.plot(df_w100['sample_size'], df_w100['plink_time_taken'], label="plink: window size 100", marker='+', color="red")
plt.plot(df_w100['sample_size'], df_w100['tskit_time_taken'], label="tskit: window size 100", marker='+', color="blue")
plt.title("LD calculation times")
plt.xlabel("sample_size")
plt.ylabel("time taken (s)")
plt.legend()
plt.show()
```
