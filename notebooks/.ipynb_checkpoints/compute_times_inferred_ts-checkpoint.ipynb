```python
num_sites = 500
max_dist = 200_000
chr = 22
ld_calc = ld_calculator.LdInfo(ts, chr)
sites_a = np.sort(np.random.choice(ld_calc.get_single_mutation_sites(), size=num_sites))
sites_b = ld_calc.return_random_sites_in_range(sites_a, max_dist=max_dist)
r2, compute_times = ld_calc.compute_ld_with_time(sites_a, sites_b)
distances = ld_calc.compute_distance_between_sites(sites_a, sites_b)
distances = np.array(distances) / 1_000
```

```python
# histogram of distances
plt.hist(np.array(distances) / 1_000)
plt.title(f"distance between queried variants")
plt.xlabel("distance (kb)")
plt.show()
```

```python
plt.hist(compute_times)
plt.title(
    f"Time taken for {num_sites} pairwise LD calculations\n(num_sequences={ts.num_samples/1_000}k, num_sites={len(ld_calc.get_single_mutation_sites()):,}, max dist considered={max_dist/1_000} kb )"
)
plt.xlabel("time taken in seconds")
plt.xlim(0, max(compute_times))
plt.show()
```

```python
# make a scatterplot of distance vs time
plt.scatter(x=np.array(distances), y=compute_times, s=5)
plt.plot(
    np.unique(distances),
    np.poly1d(np.polyfit(distances, compute_times, 1))(np.unique(distances)),
    linewidth=1,
)
plt.title(f"distance between variants v query time\n UKB 20k samples, chr22")
plt.xlabel("distance (kb)")
plt.ylabel("time taken (seconds)")
plt.text(0.3, 0.7, f"corr = {np.corrcoef(distances, compute_times)[0,1]:.2f}")
plt.show()
```
